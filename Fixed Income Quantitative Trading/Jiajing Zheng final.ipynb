# 1. Use Sample A to compute 4 cointegrated pairs of futures rates: [3y,4y], [4y,5y], [2y,4y],[3y,5y]

```python
import statsmodels.api as sm

def CCA_Chou_Ng(data_set):
    
    #data_set is pandas dataframe
    
    #define df_lag as the 1-period lag of data_set
    df_lag = data_set.shift(1).dropna()
    
    #drop the 0-index of data_set to make sure df and df_lag are aligned
    df = data_set.drop(data_set.index[0]).dropna()
    n = len(data_set.columns)
    
    #X(t) ~ M_1 + X(t-1)
    
    # input variables
    X = df_lag.values
    
    # ensures there is an intercept
    X_I = sm.add_constant(X)
    
    # regression target
    Y = df.values
    
    # fit regression
    l1 = sm.OLS(Y,X_I).fit()
    
    # get the coefficients/betas without the intercept
    B=l1.params[1:(n+1)]
    #print(B)
    
    #X(t-1) ~ M_2 + X(t)
    
    # regression, reversing the dependent and the independent
    Y_I = sm.add_constant(Y)
    l2 = sm.OLS(X,Y_I).fit()
    
    # get the coefficients/betas without the intercept
    A=l2.params[1:(n+1)]
    #print(A)
    
    # take the dot product of the two coefficient matrices
    C = np.dot(A,B)
    print(C)
    eig_val, eig_vec = np.linalg.eig(C)
    return eig_val, eig_vec, C
```

```python
import numpy as np
pair = [[3, 4], [4, 5], [2, 4], [3, 5]]
pair_dict = {}
weight_dict = {}
result_dict = {}
for i in range(len(pair)):
    pair_dict[i] = df[[f"ED{j*4}" for j in pair[i]]]
    e_val, e_vec, C = CCA_Chou_Ng(pair_dict[i].iloc[train_beg: train_end,:])
    weight_dict[i] = e_vec[:, np.argsort(e_val)[0]] / e_vec[0, np.argsort(e_val)[0]]
    result_dict[i] = pair_dict[i] @ weight_dict[i].T 
```

# 2. Use the same Sample A to construct the following: a. AR(1) model fitted to each of the 4 cointegrated pairs (Signal 1) b. AR(1) model fitted to each {cointegrated pair - EMA()} (Signal 2)

```python
from statsmodels.tsa.ar_model import AutoReg
def signal(df, lambd, train_beg, train_end, cv_beg, cv_end):
    df_train = df.iloc[train_beg: train_end]
    df2 = df - df.ewm(alpha=lambd).mean()
    df_train2 = df2.iloc[train_beg: train_end]
    
    AR_1 = AutoReg(df_train.values, lags=1)
    AR_2 = AutoReg(df_train2.values, lags=1)
    
    res1 = AR_1.fit()
    res2 = AR_2.fit()
    
    s_1 = df.iloc[cv_beg: cv_end] - AR_1.predict(params=res1.params, start=cv_beg, end=cv_end-1)
    s_2 = df2.iloc[cv_beg: cv_end] - AR_2.predict(params=res2.params, start=cv_beg, end=cv_end-1)

    return s_1, s_2
```

# 3. Compute half-lives for all signals. Pick  to make sure that half-life of Signal 2 is ~5 da

```python
def halflife(x):
    x_lag = x.shift(1)
    x_lag.iloc[0] = x_lag.iloc[1]
    x_ret = x - x_lag
    x_ret.iloc[0] = x_ret.iloc[1]

    x_lag2 = sm.add_constant(x_lag)
    model = sm.OLS(x_ret,x_lag2)
    res = model.fit()

    halflife = round(-np.log(2) / list(res.params)[1], 4)
    return halflife
```
