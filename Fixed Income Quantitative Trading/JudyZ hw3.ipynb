```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
r0 = 0.04
mu = 0.0025
k = 0.05
gamma = 0
sigma = 0.01
tau = np.linspace(0, 30, num=100)
```

```python
def funcAB(y, tau, mu, k, gamma, sigma):
    A, B = y
    return [-mu*B + 0.5*sigma*B**2, -k*B - 0.5*gamma*B**2 + 1]
```

```python
sol = odeint(funcAB, (0, 0), tau, args=(mu, k, gamma, sigma))
plt.plot(tau, sol[:, 0], 'r', label='A')
plt.plot(tau, sol[:, 1], 'b', label='B')
plt.legend(loc='best')
plt.xlabel('Time to Maturity')
plt.ylabel('Value')
```

```python
affine_price = np.exp(sol[:,0] - sol[:,1] * r0)
plt.plot(tau, affine_price, label='Affine Price')
plt.legend(loc='best')
plt.xlabel('Time to Maturity')
plt.ylabel('Price')
plt.grid()
```
