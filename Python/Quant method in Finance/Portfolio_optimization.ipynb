```python
from pandas_datareader import data as web
import time
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
equity = pd.read_csv('/Users/jiajing/Desktop/sp500.csv')
equity
```

```python
#remove <10 year stocks in S&P 500
c_d = {}
for i in range(505):
    date = equity['Date first added'][i]
    try:
        date = date[:4]
        if int(date)>2011:
            pass
        else:
            c_d[equity['Symbol'][i]]= equity['Date first added'][i]
    except:
        pass

c_d
```

```python
# form company symbols
company = list(c_d.keys())
print(company)
```

```python
#Get portfolio starting date
stock_start = '2011-12-30'
#End date
today = datetime.datetime.now().strftime('%Y-%m-%d')
today
```

```python
#Create dataframe to store the adjusted close price of the stock
df = pd.DataFrame()

try:
    for stock in company:
        df[stock] = web.DataReader(stock,'yahoo',stock_start,today)['Adj Close']
except:
    pass

df
```

```python
#return
return_stock = df.pct_change()
cov_annual = return_stock.cov()*252 
cov_annual
```

```python
#assign weights to the stock
 #number of company in df
w = 1/41
weight_l = []
for i in range(41):
    weight_l.append(w)

weights = np.array(weight_l)
weights
```

```python
#portfolio variance
p_variance = np.dot(weights.T, np.dot(cov_annual,weights))
p_variance
```

```python
#Portfolio volatility(sd)
p_volatility = np.sqrt(p_variance)
p_volatility
```

```python
#calculate annual portfolio return
p_annualreturn = np.sum(return_stock.mean()*weights) * 252
p_annualreturn
```

```python
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns
```

```python
#Optimaztion
mean = expected_returns.mean_historical_return(df)
risk = risk_models.sample_cov(df)

optimization = EfficientFrontier(mean,risk)
w = optimization.max_sharpe()
new_w = optimization.clean_weights()
print(new_w)
optimization.portfolio_performance(verbose = True)
```
